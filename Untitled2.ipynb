```python
# ==============================
# Real-Time Mental Health Risk Monitoring
# Dataset: addiction_2018_features_tfidf_256.csv
# Label: guns_total
# Models: Logistic Regression, SVM, Naive Bayes, CNN-LSTM
# Metrics: Accuracy, Precision, Recall, F1, Confusion Matrix, ROC
# ==============================

!pip install vaderSentiment --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Conv1D, LSTM
from tensorflow.keras.optimizers import Adam

# 1️⃣ Load dataset
df = pd.read_csv("addiction_2018_features_tfidf_256.csv")
print("Dataset shape:", df.shape)

# 2️⃣ Set label column
label_col = "guns_total"  # target variable

# 3️⃣ Drop non-numeric columns except label
df = df.drop(columns=[c for c in df.columns if df[c].dtype == 'object' and c != label_col])

# 4️⃣ Encode label if needed
if df[label_col].dtype == 'object':
    df[label_col] = df[label_col].astype('category').cat.codes

# Check label class count
if df[label_col].nunique() < 2:
    raise ValueError(f"Label '{label_col}' has only one class.")

# 5️⃣ Train-test split
X = df.drop(columns=[label_col])
y = df[label_col]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# ==============================
# Evaluation function
# ==============================
results = {}
def evaluate_model(name, y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    rec = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    results[name] = [acc, prec, rec, f1]
    print(f"\n{name} Results:")
    print(f"Accuracy: {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall: {rec:.4f}")
    print(f"F1 Score: {f1:.4f}")

# ==============================
# Logistic Regression
# ==============================
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)
evaluate_model("Logistic Regression", y_test, y_pred_lr)

# ==============================
# SVM
# ==============================
svm = SVC(kernel='linear', probability=True)
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)
evaluate_model("SVM", y_test, y_pred_svm)

# ==============================
# Naive Bayes
# ==============================
X_train_nb = np.maximum(X_train, 0)
X_test_nb = np.maximum(X_test, 0)
nb = MultinomialNB()
nb.fit(X_train_nb, y_train)
y_pred_nb = nb.predict(X_test_nb)
evaluate_model("Naive Bayes", y_test, y_pred_nb)

# ==============================
# CNN-LSTM
# ==============================
X_train_dl = np.expand_dims(X_train.values, axis=2)
X_test_dl = np.expand_dims(X_test.values, axis=2)

model = Sequential()
model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_dl.shape[1], 1)))
model.add(Dropout(0.3))
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(np.unique(y)), activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
model.fit(X_train_dl, y_train, epochs=10, batch_size=32, validation_data=(X_test_dl, y_test), verbose=1)

y_pred_dl = np.argmax(model.predict(X_test_dl), axis=-1)
evaluate_model("CNN-LSTM", y_test, y_pred_dl)

# ==============================
# Confusion Matrix for Best Model
# ==============================
best_model_name = max(results, key=lambda x: results[x][0])
print(f"\nBest model based on Accuracy: {best_model_name}")

cm = confusion_matrix(y_test, eval(f"y_pred_{'lr' if 'Logistic' in best_model_name else ('svm' if 'SVM' in best_model_name else 'nb' if 'Naive' in best_model_name else 'dl')}"))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title(f"Confusion Matrix - {best_model_name}")
plt.show()

# ==============================
# ROC Curve (Binary Classification)
# ==============================
if len(np.unique(y)) == 2:
    model_map = {'Logistic Regression': lr, 'SVM': svm, 'Naive Bayes': nb, 'CNN-LSTM': model}
    X_input = X_test if best_model_name != 'CNN-LSTM' else X_test_dl
    probas = model_map[best_model_name].predict_proba(X_input)[:, 1] if best_model_name != 'CNN-LSTM' else model.predict(X_input)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, probas)
    auc = roc_auc_score(y_test, probas)
    plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {best_model_name}")
    plt.legend()
    plt.show()

# ==============================
# Final Results Table
# ==============================
final_df = pd.DataFrame(results, index=["Accuracy", "Precision", "Recall", "F1"]).T
print("\nFinal Model Comparison:")
print(final_df)
```
